```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_squared_error, r2_score

from pickle import dump
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.drop(columns=['package_name'])
```

```python
df['review'] =  df['review'].str.strip().str.lower()
```

```python
X = df['review']
y = df['polarity']
```

```python
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python
model = MultinomialNB()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print(accuracy_score(y_test, y_pred))
```

Voy a probar un random forest

```python
# Entrenar el modelo Random Forest
model_rf = RandomForestClassifier(max_depth = 30, max_features = 'sqrt', min_samples_leaf = 5, min_samples_split = 15, random_state = 42)
model_rf.fit(X_train, y_train)
```

```python
# Realizar predicciones en el conjunto de entrenamiento y prueba
y_train_pred_rf = model_rf.predict(X_train)
y_test_pred_rf = model_rf.predict(X_test)

y_train_pred_rf = model_rf.predict(X_train)
y_test_pred_rf = model_rf.predict(X_test)
```

probando parametros con grid search

```python
params = {'max_depth': [10, 20, 40, 70, 100, None],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [200, 600, 1200, 1400],
 'n_jobs': [3]}
```

```python
grid_search = GridSearchCV(estimator=model_rf, param_grid=params, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
# Evaluar el modelo Random Forest
accuracy_train_rf = accuracy_score(y_train, y_train_pred_rf)
accuracy_test_rf = accuracy_score(y_test, y_test_pred_rf)

f1_score_train_rf = f1_score(y_train, y_train_pred_rf, average='micro')
f1_score_test_rf = f1_score(y_test, y_test_pred_rf, average='micro')

precision_train_rf = precision_score(y_train, y_train_pred_rf, average='micro')
precision_test_rf = precision_score(y_test, y_test_pred_rf, average='micro')

recall_train_rf = recall_score(y_train, y_train_pred_rf, average='micro')
recall_test_rf = recall_score(y_test, y_test_pred_rf, average='micro')

print("Random Forest Metrics:")
print("Accuracy Train: ", accuracy_train_rf)
print("Accuracy Test: ", accuracy_test_rf)
print("F1 score Train: ", f1_score_train_rf)
print("F1 score Test: ", f1_score_test_rf)
print("Precision Train: ", precision_train_rf)
print("Precision Test: ", precision_test_rf)
print("Recall Train: ", recall_train_rf)
print("Recall Test: ", recall_test_rf)
```
